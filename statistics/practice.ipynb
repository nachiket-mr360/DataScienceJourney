```python
import pandas as pd
import matplotlib.pyplot as plt

marks = [
    40,42,43,44,45,45,46,47,48,49,
    50,50,51,52,53,54,55
]

df = pd.DataFrame({"Marks": marks})

plt.hist(df["Marks"], bins=6)
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Marks")
plt.show()
```

```python
salary = [
    25000,28000,30000,32000,35000,
    38000,40000,42000,45000,50000,
    300000
]
df = pd.DataFrame(salary)
plt.hist(df)
```

```python
#right skewed histogram
d = {"salary": [25,30,35,40,45,50,150]}
df = pd.DataFrame(d)
plt.hist(df)
```

```python
#left skewed histogram
d = {"salary": [-20, 25,30,35,40,45,50]}
df = pd.DataFrame(d)
plt.hist(df)
```
