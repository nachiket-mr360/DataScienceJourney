```python
import seaborn as sns
import matplotlib.pyplot as plt



d = {"salary": [-10,25,30,35,40,45,50,90]}
df = pd.DataFrame(d)

df["salary"].mean()
df["salary"].median()
sns.boxplot(x=df["salary"])
plt.show()
```
