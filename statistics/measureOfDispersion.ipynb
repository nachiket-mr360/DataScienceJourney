```python
import numpy as np
```

```python
data = [2,3,4,5,8]
data
```

```python
#range
max(data) - min(data)
```

```python
#calculate percentile using numpy
np.percentile(data, [0, 25,50,75,100]) #here 25 refer to q1, 50 refer to median or q2, 75 --q3 and 100 --q4
```

```python
data = [2,3,4,5,9,9,12,100]
np.percentile(data,[0,25,50,75,100])
```

```python
#IQR = q3-q1
#we can directly by looking above
print(9.75 - 3.75)
#else
q1 = np.percentile(data,[25] )
q3 = np.percentile(data,[75])
iqr = q3-q1
print(iqr)
```

```python
#IQR is used to plot box-viscur plot to know the outliers
import seaborn as sns
sns.boxplot(data)
#here in the plot we can easily see the outlier (the data is far from opper fence)
#upper fence = q3+1.5*iqr
#lower fence = q1 - 1.5*iqr (as we know value below lower fence and value above upper fence considered as outlier)
#this can easily caught by box-viscur plot
```

```python
#variance 
np.var(data)
```

```python
np.std(data)
```

```python
data = [1,2,3,3,4,4]
np.var(data)
```

```python
np.std(data)
```

```python
#another library for knowning variance and standard deviation
import statistics as st
print(st.variance(data))
#as we compared to out original variance of numpy it's lower than it, because it is calculating for sample 
#to calculate for population we need to use pvariance
print(st.pvariance(data))
```

```python
print(st.stdev(data)) #for sample
print(st.pstdev(data)) #for population
```
