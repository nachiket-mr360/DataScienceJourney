```python
#set operations >> union, intersection, difference, superset, subset, symmentric difference
a = {32,34,67,78,100}
b = {67,78,30}
c = {67,78}

#intersection
print(a & b) #gives intersection also we can perform it using
print(a.intersection(b))
#union 
print(a | b) #gives all element present in both Or
print(a.union(b))
#difference
print(a-b)
print(b-a)
print(a.difference(b))
#subset
print(a.issubset(b))
print(b.issubset(a))
print(c.issubset(a))
#super set
print(a.issuperset(c))
print(a.issuperset(b))
#symmentic difference
print(a.symmetric_difference(b))
print(b.symmetric_difference(c)) #or
print(a^c)
```
