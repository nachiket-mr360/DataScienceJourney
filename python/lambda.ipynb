```python
#lambda funtion : ananymous function (which doesnot required to follow traditional method)
#traditional method
def even_check(num):
    if num %2 ==0:
        print('even')
    else:
        print('Odd')
even_check(2)
```

```python
#using lambda
#syntax -- lambda arguments : expression (single line)


even_check = lambda num : 'even' if num%2 ==0 else 'odd'
even_check(3)
```

```python
def square(x):
    return x**2 #explicit return funtion
square(4)
```

```python
square = lambda x: x**2 #in lambda it understand return implicitly
square(3)
```

```python
#add two numbers
addition = lambda x,y : x+y
addition(1,3)
```

```python
x= ['python','data scinece','nachiket']

sort_word = sorted(x, key = lambda x : len(x))
sort_word
```

```python
#fibbonacii
fib = lambda n: n if n<=1 else fib(n-1)+fib(n-2)
[fib(i) for i in range(10)]
```

```python
#factorial 
fact = lambda n : 1 if n==0 else n*fact(n-1)
fact(5)
```
