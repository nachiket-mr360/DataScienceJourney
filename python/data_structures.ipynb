```python
def check_subset_superset(s1, s2):
    is_subset = s1.issubset(s2)
    is_superset = s1.issuperset(s2)
    is_equal = True if s1==s2 else False
    if is_subset and is_equal !=True:
        print('Set1 is a subset of Set2')
    elif is_superset and is_equal !=True:
        print('Set1 is a superset of Set2')
    elif is_equal:
        print('Set1 and Set2 are equal')
    else:
        print('No subset or superset relation')

check_subset_superset({1,2,3},{1,2,3})
```

```python
# jewels and stones
'''you're given strings jewels representing the types of stones that are jewels, and stones representing the stones you have,
each character in stones is a type of stone you have, you 
want to know how many of the stones you have are also jewels
letter are case sensitive, so "a" is considered as different type
of stone from "A"'''
''' 
ex1: jewels = "aA" , stones ="aAAbbbb" (here 3 are jewels)
in stones we have 7 stones, a, A, A, b,b,b,b and we know a and A are jewels so
3 jewels are present in 7 stones
ex2: jewels ="z" , stones ="ZZ" (0 jewels)'''

def numJewelsInStones(jewels: str,stones:str):
        d = {}
        for i in stones:
            if i in d.keys():
                d[i] +=1
            else:
                d[i] = 1
        count = 0
        for i in d.keys(): #similar to for i in d:
            if i in jewels:
                count += d[i]
        return count
numJewelsInStones('aA','aAAbbbb')
```

```python
#sort the people
''' 
you are given an array of strings names, and ana array heights that consists of distinct
positive integers both array are of length n
for each index o, names[i] and heights[i] denote the name and height of the ith person
return names sorted in descending order by the people hights
ex:
input names= [mary, john, emma]
heights = [180,165,170]
output: mary is the tallest, followed by emma and john'''

def sort_the_people(names, heights):
    d = {}
    for i in range(len(names)):
        d[names[i]] = heights[i]
        
    d = dict(sorted(d.items(), key = lambda x:x[0], reverse=True))
    return list(d.keys())
    
sort_the_people(['nachi','akash','john'],[180,165,170])
```

```python
#check two strings have same frequency map
#campare character frequncies of two strings and check if they match
# Check whether two strings have the same character frequencies

def compare_frequency(word1, word2):

    if len(word1) != len(word2):
        return False

    freq = {}

    for char in word1:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1

    for char in word2:
        if char in freq:
            freq[char] -= 1
        else:
            return False

    for value in freq.values():
        if value != 0:
            return False

    return True


print(compare_frequency('aabbcc', 'baccab'))
print(compare_frequency('dhsnd', 'dhnas'))
```

```python
#most frequent even element
def frequent_even(numbers):
    d = {}
    
    for i in numbers:
        if i %2 ==0:
            if i in d:
                d[i] +=1
            else:
                d[i] = 1
    if not d:
        return -1
    max_f = max(d.values())
    cand = [num for num, freq in d.items() if freq == max_f]
    return min(cand)
    
frequent_even([1,4,3,2,2,4,5,4,5,6,7,2,1,7,4])
```

```python
#check if number has equal digit count and digit value
''' 
u are given a 0-indexed string num of len n consisting of digits
return true if for every index i in the range 0<=i <n the digit i occurs num[i]
times in num, otherwise return false
'''
def digitCount(num):
    d = {}
    for i in num:
        if i in d:
            d[i] += 1
        else:
            d[i] = 1
    for i in range(len(num)):
        if d.get(str(i),0) == int(num[i]):
            continue
        else:
            return False
    return True
digitCount('1210')
```

```python

def invert_dictionary(d):
    c = {}
    for key, value in d.items():
        c[value] = key
    return c
invert_dictionary({'a': 10, 'b': 25, 'c': 20})
```

```python
def find_common_keys(dict1, dict2):
    d = []

    for i in dict1:
        if i in dict2:
            d.append(i)
            
    return d

find_common_keys({'a':1,'b':2,'c':3},{'b':4,'c':5,'d':6})
```

```python
def remove_empty_or_none_values(d):
    
    j = {}
    for key, value in d.items():
            if value:
                j[key] = value
    return j
remove_empty_or_none_values({'x': '', 'y': 20, 'z': None})
```

```python
def create_dict_from_lists(keys, values):

    # Find smaller length
    min_length = min(len(keys), len(values))

    result = {}

    for i in range(min_length):
        result[keys[i]] = values[i]

    return result


create_dict_from_lists(['a', 'b', 'c'],[1, 2, 3])
```

```python
def keys_with_even_values(data):
    d = []

    for i in data:
        if data[i]% 2 ==0:
            d.append(i)
    return d
keys_with_even_values({'a': 2, 'b': 3, 'c': 4})
```

```python
def merge_list_of_dicts(dict_list):
    
    d = {}

    for i in dict_list:
        for j in i:
            if j not in d:
                d[j] = i[j]
            else:
                d[j] = i[j]
    return d

merge_list_of_dicts([{'a': 1, 'b': 2}, {'b': 3, 'c': 4}])
```

```python
def group_values_by_key(dict_list):
    d = {}

    for i in dict_list:
        for key, value in i.items():
            if key in d:
                d[key].append(value)
            else:
                d[key] = [value]
    return d
group_values_by_key([{'a': 1, 'b': 2}, {'b': 3, 'c': 4}])
```

```python
def print_unique_elements(numbers):
    d ={}
    unique = []
    for i in numbers:
        if i in d:
            d[i] +=1
        else:
            d[i] = 1
    for i in d:
        if d[i] == 1:
            unique.append(i)

    return unique
print_unique_elements([3,4,5,2,3,1])
```

```python
def frequency_count_tuple( elements):
    d ={}

    for i in elements:
        if i in d:
            d[i] +=1
        else:
            d[i] = 1
    return d

frequency_count_tuple((1,2,3,4,6,6,3,4,4))
```

```python
def remove_duplicates_using_set( elements):
    
    s= set(elements)
    return sorted(list(s))
remove_duplicates_using_set([10,20,10,30, 40, 20])
```

```python
def sort_dict_by_values(data):
    d = {}
    for key, value in data.items():
        d[key]= value
    
    j = sorted(d.items(), key=lambda  item:item[1])
    return dict(j)
sort_dict_by_values({'a': 3, 'b': 1, 'c': 2})
```

```python
def find_all_sublists(lst):

    result = []

    for i in range(len(lst)):

        for j in range(i + 1, len(lst) + 1):

            result.append(lst[i:j])

    return result

print(find_all_sublists([1, 2, 3]))
```
