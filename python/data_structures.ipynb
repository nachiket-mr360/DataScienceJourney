```python
#linear search 

li = list(map(int, input().split()))
key = int(input("enter search key: "))

for i in range(len(li)):
    if li[i] == key:
        print(f'key found at {i} index')
        break
else:
    print('key not found')
#linear search has time complecity t(o)
```

```python
#binary search
#efficiently search in sorted list, using divide and conque
li = list(map(int, input("enter list element in sorted order").split()))

for i in range(len(li)-1):
    if li[i] < li[i+1]:
        continue
    else:
        print("list is not sorted you can't perform binary search")
        break
else:
    print('list is sorted, you can perform binary search ')
    key = int(input('enter search key: '))
    
    start = 0
    end = len(li)-1
    mid = (start+end)//2
    
    while start <= end:
        if li[mid] == key:
            print(f'element found at {mid} index')
            break
        elif li[mid] > key:
            end = mid-1
            mid = (start+end)//2
        elif li[mid] < key:
            start = mid+1
            mid = (start+end)//2
    else:
        print('no such element exists')
            
```

```python
#bubble sort

li = list(map(int, input().split()))

for i in range(len(li)-1):
    for j in range(len(li)-1-i):
        if li[j] >li[j+1]:
            li[j], li[j+1] = li[j+1], li[j]
print(li)
```

```python
#replace even number with zero
def replace_even_with_zero(lst):
    # Write your code here
    for i in range(len(lst)):
        if lst[i] %2 ==0:
            lst[i] =0
    return lst
replace_even_with_zero([3,3,5,2,6,3,2])
```

```python
#slit list into 3 halfs , if odd then len(list1) > len(list2)
def split_list_in_halves(numbers):
    # Write your code here

    mid = (len(numbers)+1)//2
    first_half = numbers[:mid]
    second_half = numbers[mid:]

    return first_half, second_half

split_list_in_halves([4,5,2,5,1])
```

```python

def remove_specific_element(numbers, x):
    # Write your code here
    for i in range(len(numbers)):
        if numbers[i] == x:
            numbers.remove(x)
            
            return numbers
    else:
        return numbers
remove_specific_element([2,3,4,5],1)
remove_specific_element([2,4,5,2,4],2)
```
