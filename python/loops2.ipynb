```python
#loops : loops are fundamental part of programing
#it allows us to repeatedly use a part of code, based on a condition
```

```python
#while loop
#it execute repeatedly a block of code as long as a condition is true
n = 5
i = 1
while i<n:
    print(i)
    i +=1
```

```python
count = 5
while count>0:
    print(count)
    count -=1
```

```python
n = 7
i = 1
while i<n:
    print(i) 
    i +=1
else:
    print('this will be executed when the while loops is run succesfully')
```

```python
n = 7
i = 1
while i<n:
    print(i) 
    i +=1
    if i ==3:
        break #this will break the loop and when break is used the else part also not executed
else:
    print('this will be executed when the while loops is run succesfully')
    
```

```python
#continue : used to skip the iteration 
n = 7
i = 1
while i<n:
     
    i +=1
    if i ==3:
        continue #this will only skip the 3 and continue with 4 and also execute the else block
    print(i) 

else:
    print('this will be executed when the while loops is run succesfully')
    
```

```python
'''
Take a number from user.
Print:

"Positive Even"
"Positive Odd"
"Negative Even"
"Negative Odd"
"Zero" 
'''
```

```python
n = int(input("Enter the number: "))
if n <0:
    if n %2 == 0:
        print('negative even')
    else:
        print('negative odd')
elif n>0:
    if n%2 == 0:
        print('positive even' )
    else:
        print('positive odd')
else:
    print('zero')
    
```

```python
'''Q3.

Take a number n.
Print sum of all numbers from 1 to n.
'''
n = int(input("enter nth term: "))
total = 0
for i in range(n+1):
    total +=i
print(total)
```

```python
'''Take a number from user and check whether it is:

Prime or Not Prime'''
n = int(input("enter number: "))
is_prime = True
if n <=1:
    is_prime = False
else:
    for i in range(2, int(n**0.5)):
        if n %i ==0:
            is_prime = False
            break
if is_prime:
    print('prime')
else:
    print('not prime')
```

```python
def factorial(n):
    if n<=1:
        return 1
    else:
        return n * factorial(n-1)
        
factorial(5)
```

```python
#count vowels
def count_vowels(text):
    count = 0
    for i in text.lower():
        if i in 'aeiou':
            count +=1
    return count
count_vowels('nachiket')
```
