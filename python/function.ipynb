```python
#funtion for palindrome
def palindrome(n):
    rev = 0
    temp = n
    while n>0:
        rev = (rev*10) + (n%10)
        n //= 10
    if rev == temp:
        return True
    else:
        return False
    
print(palindrome(121))
print(palindrome(200))
print(palindrome(323))
```

```python
def sum_of_naturals(n):
    # Write your code here
    total = 0
    if n <=0:
        print(total)
    else:
        for i in range(1,n+1):
            total = total + i
        print(total)

sum_of_naturals(12)
```

```python
def reverse_number(n):
    # Write your code here
    sign = -1 if n<0 else 1
    n = abs(n)
    reversed =0
    while n>0:
        reversed = reversed*10 + n %10
        n //= 10
    print(reversed*sign)
reverse_number(345)
```

```python
def recursive_sum(n):
    if n == 0:
        return 0
    return n + recursive_sum(n-1)

recursive_sum(12)
```

```python
def inverted_triangle(n):
    for i in range(n,-1,-1):
        for j in range(i):
            print('*',end='')
        print()
inverted_triangle(5)
```

```python
#take two list of same lenth, and create a new list which is sum of the corresponding elements from both lists
def element_wise_sum(list1, list2):
    list3 = []
    if len(list1) == len(list2):
        for i in range(len(list1)):
            list3.append(list1[i]+list2[i])
    print(*list3)
element_wise_sum([1,2,3],[4,5,6])
```
