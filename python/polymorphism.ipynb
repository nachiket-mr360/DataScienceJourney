```python
'''
Polymorphism 
polymorphism allows different classes to define methods with the same name but
different behaviors in python , it is typically achieved through method overriding

types of polymorphism
method overloading: python simulates method overloading using default or variable-length arguments
as it doesnot support traditional overloading

method overriding: occurs when a subclass defines a method with the same name as its superclass,
replacing the superclass's method'''

def hello():
    print('hello i am a func')
    
def hello():
    print('hello i am 2nd function') #it redefined the old funtion so when u call hello() it gives this instead of old one
    
hello()
```

```python
#when we want to use both 
class Animal:
    name = 'lion'
    def speak(self):
        print('hello i roar')

class Bird:
    name = 'sparrow'
    
    def speak(self):
        print('hello i tweet')

#there are two same function called speak, but they have different functional activity
obj = Animal()
obj2 = Bird()

obj.speak() #here is roar
obj2.speak() #here it tweet
```

```python
class Animal:
    
    def speak(self):
        print('hello i roar')
        
class Human(Animal):
    name = 'akarsh'
    def speak(self):
        super().speak() # we can also called parent method in that
        print('hello my name is nachiket')

obj = Human()
obj.speak()
obj2 = Animal()
obj2.speak()
```

```python
'''   
Abstraction 
abstraction is used to simplifying complex systems by focusing on essential
features and hiding unnecessary details
it is used to define a common interface for different sub classes

abstract classes and methods
abstract class : a class that contains one or more abstract methods

abstract method: a method that is defined but not implemented in the abstract class
subclasses must provide the implementation

'''

from abc import ABC, abstractmethod  #to use abstractmethod we need to import it from abc library

class Animal(ABC): #now the animal class is becomes abstract
    @abstractmethod
    def sound(self):
        pass
    
class Dog(Animal):
    def hello(self):
        print("it's a dog")

obj = Dog()    

#it will give an error that 
#TypeError: Can't instantiate abstract class Dog without an implementation for abstract method 'sound'
#because we can't call dog without using sound method inside dog which is rule for abstract method
```

```python

from abc import ABC, abstractmethod  
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass
    
class Dog(Animal):
    def hello(self):
        print("it's a dog")
    def sound(self):
        print('hello i make woff sound')
        return super().sound()

obj = Dog()    #now we will not get error because we used sound method in abstract class
obj.sound()
obj.hello()
```
