```python
#leap year checker (rules : divide by4, not by 100 unless divisible by 400)
year = int(input("enter the year: "))
if (year %4 == 100 and year %400 ==0) or (year %4 ==0 and year%100 !=0):
    print("leap year")
else:
    print("not a leap year")
```

```python
#shop discount calculator
'''ask for purchase amount, apply discounts based on threshould, ex: above 1000-> 10%
off, above 5000-> 20% off, print final bill, '''
bill = int(input("enter bill amount: "))
if bill >=1000 and bill <=4999:
    print(f'you got a discount of 10% your final bill amount is: {(bill*90)/100}')
elif bill >=5000:
    print(f'you gor a discount of 20% of your bill, final bill is {(bill*80)/100}')
else:
    print("sorry no discount for you, final bill is {bill}")
```

```python
# write your code here
number = int(input())
sign = -1 if number <0 else 1
temp = abs(number)
reverse =0
while temp >0:
    last_dig = temp %10
    reverse = reverse*10 + last_dig
    temp //=10
print(reverse*sign)
```
