```python
#parameters
#modes in files 
''' 
r = read mode (by default mode)
x = creating file (error if already exist) or w also used to create file
(overwrite if present)
a = for appending file
a+ = appending and reading, r+ = reading and writing (without truncating)
w+ = reading and writing with truncate
'''

file3 = open('test for w', 'w')

file3.write('hello this is a sample file that i have created')

file3.close()

file3 = open('test for w')
file3.read()
```

```python
file3 = open('test for w','a')
file3.write('new line added\n in file')
file3.close()
```

```python
file3 = open('test for w')
file3.read()
```

```python
#to avoid creating variable again and again we can use with (And we also avoid to use close() function)

with open("test for w", 'r') as fs:
    print(fs.read())
```

```python
with open('file2.txt','r+') as fs:
    fs.write('hello new file is created')
    print(fs.read())
```
