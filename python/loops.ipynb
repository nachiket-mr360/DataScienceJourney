```python
#Automorphic number 
#a number is automorphic if its square ends with the number itself(eg 5^2 =25, 76^2 = 5776)
n = int(input("Enter the number: "))
temp = n
square = n ** 2
count = 0
while n>0:
    n //=10
    count +=1
extracted = square % (10 ** count)
if extracted == temp :
    print('your number is automorphic')
else:
    print("not an automorphi")
```

```python
#check palindrome numbers in a given range of two numbers and prime them
start, end = map (int, input().split())
found = 0
for number in range(start, end+1):
    temp = number
    reversed = 0

    while temp > 0:
        dig = temp %10
        reversed = reversed * 10 + dig
        temp //=10
    if reversed == number:
        found +=1
        print(number, end=' ')
        
if found == 0:
    print('No palindrome numbers')
```

```python
word1 = input()
word2 = input()

if len(word1) != len(word2):
    print("Not Anagrams")

else:

    is_anagram = True

    for ch in word1:

        count1 = 0
        count2 = 0

        for i in word1:
            if i == ch:
                count1 += 1

        for j in word2:
            if j == ch:
                count2 += 1

        if count1 != count2:
            is_anagram = False
            break

    if is_anagram:
        print("Anagrams")

    else:
        print("Not Anagrams")
```

```python
#write a program to check strong numbers
n = int(input())
fact_sum = 0
temp = n
while n>0:
    last_dig = n %10
    fact = 1
    for i in range(1,last_dig+1):
        fact = fact*i
    n //= 10
    fact_sum = fact_sum + fact
if temp == fact_sum:
    print('Strong Number')
else:
    print("Not Strong Number")
```
