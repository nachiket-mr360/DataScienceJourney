```python
#type conversion : used to convert one datatype into another
'''there are two ways of type conversion
1.implicit type conversion (done by interpeter) (no loss of data)
2.explicit type conversion (done by ourself) (may sometimes loss of data)'''
```

```python
a = 12
b = 10.2
c = a *b #implicit conversion float
print(c)
print(int(c)) #explicit conversion float to int, (loss of data)
```

```python
#boolean
a = 23
b = 'good bye'
print(bool(a))
print(bool(b))
#there are only 7 false value otherthen those all are almost true those are
#0 , empty list,tuple, disctionary, string, false, 0.0
c=""
print(bool(c))
```
