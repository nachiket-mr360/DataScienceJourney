Naming conventions in variables
There are 3 types of naming conventions 
1.camel case -- helloWorld (use first cap in words expect first)
2.pascal case --HelloWorld (use cap for all first latters)
3.snake case -- hello_world (use underscore)
we store datatypes in variables 


```python
#camel case
nameOfPerson = input("Enter name: ")
#pascal case
AgeOfPerson = int(input("enter age: "))
#snake case 
sure_name = input("enter your sure name: ")
print(f'name: {nameOfPerson} /n age:{AgeOfPerson}/nsure name: {sure_name}')
```

```python
#accessing string elements
'''we use indexing to access string elements
string indexing starts from 0 and goes size-1 and strings even have negative indexing
like -1 denotes last element of string
ex:'''
print(a[0]) #first element
print(a[3]) # 4nd element
print(a[5]) #last element
print(a[-1]) #last element
print(a[-2]) #2nd last element
```

```python
#string slicing : we can take a portion of string
#use variable[starting point:end point:steps] #starting inclusive and end exclusive
word = 'string slicing practice'
print(word[1:5]) #gives trin
print(word[0:10:2]) #gives srn l
#we if not mention those point then default will be there
#for start point it's 0, end point (size-1) and step is 1
print(word[0:])
print(word[:14])
print(word[::])
```
