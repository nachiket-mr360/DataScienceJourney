```python
'''Dunder/magic methods
dunder methods are special methods in python that define object behavior for built-in operations
they are prefixed with double underscores 
(eg: __init__, __str__)

common dunder /magic methods
__init__ (self, ..) : constructor, called when an object is created

__add__(self, other): defines behaviour for the + operator

__eq__ (self, other): defines behavior for the == operators
__getitem__(self,key): allows indexing like lists dictionaries

'''

class Students:
    def __init__(self, name, marks):
        self.name = name
        self.marks =  marks
    def __str__(self):
        return f'name is {self.name} and marks is {self.marks}' #using this dunder method we can directly print all thing 
    #without calling perticular method
        
obj = Students('nachiket',91)
print(obj)
```

```python
class Shopping:
    def __init__(self, items):
        self.items = items
        
    def __len__(self):
        return len(self.items)
    
obj = Shopping(['apple','milk','bread','etc'])
obj2 = Shopping(['milk','curd','abc'])

print(len(obj))
```

```python
class Numbers:
    def __init__(self, number):
        self.number = number
        
    def result(self):
        pass
    def __add__(self, other):
        return self.number + other.number

obj1 = Numbers(12)
obj2 = Numbers(34)

print(obj1 + obj2)

```
