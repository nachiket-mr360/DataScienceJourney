```python
#errors: those are unexcepted behaviour of program which break the execution
# when we get error the statments after error will not executes
# there are mainly 2 types of errors
''' 
1.compile time error (syntax)
2.runtime error (like zero dision error, index out of bound etc)

exception handling are used to avoid those error and handle those error and continue the flow
of program
you cannot handle syntax error but if there is no syntax you can handle them'''

#exception handling funtion
''' 
try : this will catch the exception if any and pass to 
except : except will deal with the exception you have custom exceptions or universal
catcher/
else: else will be executed if no execption occurs or it wont be executed
finally: this will definitely run no matter what happens
raise : raises a custom error as you need. ex:'''
a = int(input("provide your number: "))
b = int(input("enter number 2: "))

try:
    print(a/b) #first it try this, if no errors then executed else goes to except block
except Exception as err: #Exception is universal exception handle all type of exception, err is alias of exception
    print(f'sorry an error {err} occured') #print if exception occured
print(a+b) #if error occur normally without try except it will not excuted but using try except it will 
```
