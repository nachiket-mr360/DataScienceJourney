```python
#comparision/relational operators (includes ==,!=, <,>,<=,>=)
#the comparision operators are used to compare two variables 
#the result of comparision operators are always boolean (in true or false)
print(5<10)
print(9>3)
print(5==4)
print(5!=4)
print(5<=3)
print(5>=3)
print((12 ==12)==True)
```
