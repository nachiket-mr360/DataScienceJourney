```python
def fibbo(n):
    sequence = []
    if n<0:
        return 'n must be a non negative'
    elif n==0:
        return 0
    elif n==1:
        return 1

    a,b = 0,1
    for _ in range(2, n+1):
        sequence.append(a)
        a,b = b, a+b
    return sequence
fibbo(5)
```

```python
#Check whether string is palindrome.
def palindrome(word):
    reversed = ''
    for i in range(len(word)-1,-1,-1):
        reversed += word[i]
    return 'palindrome' if reversed == word else 'nor palindrome'
palindrome('string')
palindrome('madam')
```

```python
#Find largest element in list WITHOUT max().
def large_and_2nd_large(lst):
    largest = lst[0]
    second_large = lst[0]
    
    for i in lst:
        if i > largest:
            second_large = largest
            largest = i
        elif i>second_large and i != largest:
            second_large = i
    return largest, second_large
large_and_2nd_large([1,3,45,3,2,4,4,33,2])
```

```python
def average_even(numbers):
    even = []
    average = 0
    count = 0
    total = 0
    for i in numbers:
        if i %2 == 0:
            even.append(i)
    for j in even:
        count +=1
        total += j
    average = total/count
    return average
average_even([10, 7, 20, 5, 30])
```
