```python
def fibbo(n):
    sequence = []
    if n<0:
        return 'n must be a non negative'
    elif n==0:
        return 0
    elif n==1:
        return 1

    a,b = 0,1
    for _ in range(2, n+1):
        sequence.append(a)
        a,b = b, a+b
    return sequence
fibbo(5)
```

```python
def second_large(num):
    first_large = num[0]
    sec_large = num[1]
    
    for i in num:
        if i> first_large:
            sec_large = first_large
            first_large = i
        elif i> sec_large and i != first_large:
            sec_large = i
    if first_large == sec_large:
            return sec_large
        
    return sec_large
second_large([2,3,4,5,2,45,9])
second_large([3,3,4,5,6])
```

```python
def remove_duplicates(num):
    s = set(num)
    return list(s)
remove_duplicates([3,4,5,3,4])
```

```python
def frequency_of_char(word):
    d = {}
    
    for i in word:
        if i in d:
            d[i] +=1
        else:
            d[i] = 1
    return d
frequency_of_char('nachiket')
```
