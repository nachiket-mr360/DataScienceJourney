```python
#instance in method 
'''instance attribute 
a attribute created using the self keyword or say an attribute created using self keyword like
self.name, self.age etc

class attribute 
attribute created inside a class without using the self keyword

instance method
a method created using the self keyword as parameter

class method 
class methods use the @classmethod decorator and do not rely on instance specific data

satic method 
a static method is a method inside a class that doesnot depend on any specific instance of the class(unlike
instance methods, which use self,)it is defined using the @staticmethod decorator and 
doesnot take the self or class parameter as its first argument
'''
class Animal:
    name = 'lion' #attribute
    def speak():  #methods 
        print('lion is roaring')
        
obj = Animal()  #here obj is instance of the class
```

```python
class Animal:
    gender = 'male' #class attribute (attribute)
    def __init__(self, name, age):
        self.name = name #instance attribute because they are created using obj (or instance)
        self.age = age
    def info(self):
        print('this is a method')
    
    @classmethod
    def clmethod(cls):  #we know that self was targeting the location of object but cls is targeting 
        #the class location
        print(f'{cls.gender} is your gender')
    
    @staticmethod    
    def hello(): #staitc method we don't need to pass any reference here (you may pass parameter if you want in function but it treated as funtion paramters)
        print('hello i am static method no requirement of any parameters')
    
    
obj = Animal('lison',12) 
obj.clmethod()
obj.hello() #static method doesnot require any location (parameters), we need to mention the @statimethod above it
```

```python
#inheritance 
'''inheritance is the mechanism by which a class (child) can use the properties
and methods of another class (parent)

types of inheritance
single-level inheritance
one class inherits from another 
    parent class to child class
    
multilevel inheritance
a class inherits from a class that is already a child of another class
    parent class -- child class -- child class
    
multiple inheritance 
a class inherits from multiple parent 
    parent class 
    parent class   > child class
    
hierarchical inheritance 
multiple classes inherit from a single parent 
    parent class < child class
                    child class
                    
'''

class Animal:  #parent class or super class
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def info(self):
        print(f'Animal name is {self.name} and age is {self.age}')
    
obj1 = Animal('lion',12)

class Memals(Animal):  #human class inherits animal  (child class or sub class)
    pass

obj2 = Animal('tiger',20)
obj3 = Memals('xyz',40) #we were not writted anything inside memals class but we inherited animal class
#ie. it's taking 2 paremeters and providing information (accessing properties of animal class)
    
obj3.info()
```

```python
#what if the child class has it own properties and methods?

class Animal:  #parent class or super class
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def info(self):
        print(f'Animal name is {self.name} and age is {self.age}')
    
obj1 = Animal('lion',12)

class Memals(Animal): 
    def __init__(self, name, age, number, group):
        super().__init__(name, age)  #whenever we call a super funtion it wil call the super class
        self.number = number
        self.group = group
        #no need to pass for name ang age because already passed in parent class
    
    

obj2 = Animal('tiger',20)
obj3 = Memals('xyz',40, 48593930,'b+')
obj3.info()  #but we can't access number and group here , so we need to create a new method to show them
```

```python
#Encapsulation
'''Encapsulation means bundling data (attributes) and method into oe unit (class)

it also involves controling access to these attributes using access modifiers

Access modifiers 
public : accessible anywhere , until now were using public attributes and methods

protected: accessible in the class and subclasses (use single underscore (_)), the protected access modifier in python
doesnot funtion as strickly as in some other languages like java or c++

private : accessible only in class 
we use __ (double underscore) to use private attributes and methods private means 
private no one can now access these methods and attributes not even child classes
'''

class Animal:
    name = 'lion'
    
    def speak(self):
        print('hello i will roar')
        
obj = Animal()

obj.name = 'zebra' #we can also change things using instance of class
print(obj.name)
print(obj.speak())

```

```python
#protected classes
class Animal:
    _name = 'lion'
    
    def _speak(self):
        print('hello i will roar')
obj = Animal()

obj._name = 'zebra'
print(obj._name)
obj._speak()
```

```python
#private access modifiers
class Animal:
    __name= 'lion' #private attributes
    
    def __speak(self): #private methods
        print('lion roaring')
        
obj = Animal()
print(obj.__name)
#we will get error AttributeError: 'Animal' object has no attribute '__name'
#because object doesnot have access for private classes
```
