```python
#different type of message
import logging

logging.basicConfig(filename='test.log' , level=logging.INFO , format='%(asctime)s') 
logging.info('this is normal information about software run')
logging.error('this is error msg')
logging.warning('this is my warnings')
logging.debug('this is my msg for debugging an event')
logging.critical('this is critical msg')
logging.shutdown()
```

```python
#hierarichy of messages/differnt level of logging
'''
1.debug > debugging msg and variables
2.info > the things are working as expected
3.warning > potential issues
4.error > serious problem
5.critical >termination of program
'''
#while creating software sometime we required information logging, sometimes debug msg
#warning and error and critical accordinly
```

```python
import logging
logging.basicConfig(filename='program.log', level=logging.DEBUG , format= '%(asctime)s %(levelname)s %(message)s', force=True)
#create a two list from given list
l = [1,'hello',[2,'world'],3,['python',4]]
#write a program to seperate integer and strings
li_int = []
li_str = []
for i in l:
    logging.info(f'processing each element {i}')
    
    if type(i) == list:
        for j in i:
            logging.info(f'proccessing sublist element {j}')
            
            if type(j) == int:
                li_int.append(j)
    elif type(i) == int:
        li_int.append(i)
        
    else:
        li_str.append(i)
logging.info(f'the final result is : {li_int} and {li_str}')
```
