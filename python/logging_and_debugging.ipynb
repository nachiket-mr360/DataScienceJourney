```python
import logging
logging.basicConfig(filename='test.log',level=logging.INFO) 
#to create a logging file, we use logging.basicConfig and give file name and level of logging
#logging.INFO used to get simple information about loginng
#when we execute it a test.log file creates
```

```python
#different type of message
import logging

logging.basicConfig(filename='test.log' , level=logging.INFO) 
logging.info('this is normal information about software run')
logging.error('this is error msg')
logging.warning('this is my warnings')
logging.debug('this is my msg for debugging an event')
logging.critical('this is critical msg')
logging.shutdown()
```

```python
#hierarichy of messages/differnt level of logging
'''
1.debug > debugging msg and variables
2.info > the things are working as expected
3.warning > potential issues
4.error > serious problem
5.critical >termination of program
'''
#while creating software sometime we required information logging, sometimes debug msg
#warning and error and critical accordinly
```

```python
import logging
logging.basicConfig(filename='test.log',level=logging.INFO) 
#when we initializing any level, the below of that level things will execute like
#in above we use info level, so below the info level like warning, error and critical execute but
#debug will not
logging.info('this is normal information about software run')
logging.error('this is error msg')
logging.warning('this is my warnings')
logging.debug('this is my msg for debugging an event') #as seen this will not execute
logging.critical('this is critical msg')
logging.shutdown()
```

```python
#similar ly
import logging
logging.basicConfig(filename="test2.log", level=logging.DEBUG, format='%(asctime)s')
logging.info('this is normal information about software run')
logging.error('this is error msg')
logging.warning('this is my warnings')
logging.debug('this is my msg for debugging an event') #as seen this will not execute
logging.critical('this is critical msg')
logging.shutdown()
```
