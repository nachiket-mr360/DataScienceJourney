```python
'''control flow statement:
control flow statmenets are statements which control the flow of program execution, 
this include many statements like conditional statement, looping statements, functions etc
'''
''' 
if else statement: used to control flow of program
it's a conditional statement which execute one part of code when condition is true and another block if condition false
like if person is greaterthan 17 then he can vote else not applicable for vote'''
age = int(input("enter your age: "))
if age>=18:
    print("you are eligible for vote")
else:
    print("your not eligible for vote")
#only one among them is executed based on condition
#pass keyword : pass is used to skip the block of code (when we dont want to write anything inside the code)
#if we leave it empty we get error to avoid that error we use pass
if age>4:
    pass
else:
    pass
```

```python
#ternary operator
print("vote") if age>=18 else print("not vote")
```

```python
#elif statement : when we want to execute the code based on multiple condition then we use elif statment
#to buy something based on money
money = int(input("enter the given amount: "))
if money <=100:
    print("i will have a ice cream")
elif money <= 200:
    print("i will have nuddles")
elif money <=400:
    print('i will have biriyani')
else:
    print("have meals in resto")
```

```python
#logical operators with elif statement
#to find the largest among 3
a=int(input("enter a"))
b=int(input("enter b"))
c = int(input("enter c"))
if a>b and a>c:
    print('a is largest')
elif b>a and b>c:
    print("b is largest")
else:
    print("c is largest")
```
