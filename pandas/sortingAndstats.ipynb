```python
import pandas as pd
import numpy as np
```

```python
data = {
    "product":['pen','notebook','pencil','marker','earser','sharpener'],
    'price':[10,50,5,30,8,12],
    'stock':[100,100,200,90,150,120],
    'rating':[4.5,4.6,4.8,4.3,4.4,3.8]
}
df = pd.DataFrame(data)
df
```

```python
df.sort_values(by='price',ascending=False)
```

```python
df.sort_values(by='price')
```

```python
df.sort_values(by='product', ascending=False)
```

```python
#sort multiple columns
df.sort_values(by=['price','stock'], ascending=[True,False])
#first priority is price then stock
```

```python
#descriptive statistics 
df.mean(numeric_only=True)
```

```python
#or
df.select_dtypes('number').mean()
```

```python
df.median(numeric_only=True)
```

```python
df.std(numeric_only=True)
```

```python
df.describe()
```

```python
#value counts show frequency in a series
df['product'].value_counts()
```

```python
df['product'].value_counts(normalize=True) #normalize =True used to get percentage
```

```python
#correlation and covariance
df.corr(numeric_only=True)
```

```python
df.cov(numeric_only=True)
```
