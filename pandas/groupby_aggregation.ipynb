```python
import pandas as pd

# New dataset: Employees, their Department, and Salary
data = {
    "Employee": ["John", "Sarah", "Mike", "Anna", "Tom", "Laura", "Steve", "Kate"],
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 70000, 72000, 65000, 60000, 52000, 75000, 58000]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
df
```

```python
df['Department'].value_counts()
```

```python
#what is the average salary of each department
#to solve this question if we do manually
print(df[(df['Department']=='IT')]['Salary'].mean())
print(df[df['Department'] =='HR'].Salary.mean())
print(df[(df['Department'] =='Finance')]['Salary'].mean())
```

```python
#it's too length to avoid this we have a funtion in pandas called groupby
df.groupby('Department')['Salary'] #without aggregation we only get the object with aggregation we get actual values
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Department')['Employee'].count()
```

```python
df.groupby('Department')['Salary'].agg(['mean','count','min','max','std'])
#when we want multiple aggragation function we we agg and provide function in list
```

```python
'''Mini Practice
Create a DataFrame with columns: Department, Employee, Salary, Age, City.
Group by Department to get average Salary and max Age.
Apply multiple aggregations using agg().
Create a pivot_table to see mean Salary by Department and City.
Build a crosstab for Department vs City.'''
```

```python
#multiple aggregations on different columns
import pandas as pd

# Same dataset
data = {
    "Employee": ["John", "Sarah", "Mike", "Anna", "Tom", "Laura", "Steve", "Kate"],
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 70000, 72000, 65000, 60000, 52000, 75000, 58000],
    "Age": [28, 32, 30, 27, 35, 29, 31, 26],
    'Cities':['mumbai','dehli','pune','raipur','mumbai','pune','pune','surat']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
df
```

```python
df.groupby('Department').agg({
    'Salary':['mean'],
    'Age':'max'
})
```

```python
df.groupby('Cities')['Salary'].agg(['max','min','mean', 'count'])
```

```python
df.pivot_table(values=['Salary'], columns=['Department','Cities'],aggfunc=['mean'])
```

```python
pd.crosstab(df['Department'], df['Cities'])
```
