```python
#data selection and filtering
#boolean indexing and conditional selection 
'''use comparison operators to build conditions
pass the condition inside df[..] to get matching rows
combine multiple conditions with & (and) | (or) and ~ (not).
parentheses are required when combining conditions'''
```

```python
import pandas as pd
import numpy as np
data = {
    "Name": ["Alice", "Bob", "Chrlie", "David", "Ali", None],
    "Age": ["25", "30", None, "40", "25", "30"],     
    "City": ["Delhi", "Mumbai", "Pune", "Bangalore", "Delhi", "Mumbai"],  
    "Score": [85, None, 78, 88, 85, 91],             
}
df = pd.DataFrame(data)
```

```python
df['Age'] = df['Age'].astype('Int64')
```

```python
df['Age']>30
```

```python
df[df['Age']>30]
```

```python
df[(df['Age']>25)&(df['Score']>80)]
```

```python
df[(df['Name'].str.startswith('A') | (df['City'].str.startswith('B')))]
```

```python
df[(df['Score']>85)]
```

```python
df[~(df['Score']>85)]
```

```python
#using query() for cleaner filtering
'''query() allows filtering rows with a string expression
it;s concise and readable for complex conditions
variables can be passed using the @ symbol
'''
#simple query : age>25
df.query('Age>25')
```

```python
df.query("Score>60")
```

```python
#multiple condition with query()
df.query("Age<35 & Score<90")
```

```python
df.query("Age>30 | City=='Delhi'")
```

```python
df.query("Age>30 or City=='Delhi'")
```

```python
hold = 90
df.query("Score >@hold") #we can use variable using @ in query
```
